## Определить корни

```python
from sympy import Symbol
from sympy import sqrt, solve, solveset, diff, plot, is_increasing, Interval, oo, S
from sympy import sin, cos, pi, simplify, periodicity, evalf
from sympy import minimum, maximum 
```

```python
x = Symbol('x', real = True)
y = simplify((sin(x)/2 + cos(x)/2)**2 * sin(x))
y
```

```python
solveset(y, x)
```

```python
solve(y, x)
```

```python
periodicity(y, x)
```

## Найти интервалы, на которых функция возрастает

```python
derivate = diff(y)
derivate
```

```python
grafic_derivate = plot(derivate, (x, -pi, pi))
```

### На графике интервалы возрастания функции расположены в интервалах положительного значения производной функции

```python
solveset(derivate>=0, x)
```

```python
solve(derivate, x, Interval (-pi, pi))
```

## Найти интервалы, на которых функция убывает

```python
solveset(derivate < 0, x)
```

### На графике интервалы убывания функции расположены в интервалах отрицательного значения производной функции

## Построить график

```python
grafic = plot(y, (x, -pi, pi))
```

## Вычислить вершину

```python
y_max = maximum(y, x, Interval (-pi, pi)).evalf()
y_min = minimum(y, x, Interval (-pi, pi)).evalf()

print('y_max =', round(y_max, 3))
print('y_min =', round(y_min, 3))
```

## Определить промежутки, на котором f > 0

```python
solveset(y>0, x, S.Reals)
```

## Определить промежутки, на котором f<0

```python
solveset(y< 0, x, S.Reals)
```
